```python
df['Arrival Delay in Minutes']=df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean)
```

```python
df.isna().sum()
```

```python
categorical_columns=['Gender','Customer Type','Type of Travel','Class','satisfaction']

for columns in categorical_columns:
    print(f'unique values in {columns} are:{df[columns].unique()}')
```

```python
sns.countplot(data=df,x='Customer Type')
plt.show()
```

```python
sns.histplot(data=df,x="Class",bins=3,kde=True)
plt.show()
```

```python
def con(x):
    if x == 'neutral or dissatisfied':
        return 0
    else:
        return 1
```

```python
df['satisfaction']=df['satisfaction'].apply(con)
```

```python
x=df.drop(columns=['satisfaction'])
x
```

```python
y=df['satisfaction']
y
```

```python
x.shape
```

```python
y.shape
```

```python
x.drop(columns='Age',inplace=True)
```

```python
x=pd.get_dummies(df,drop_first=True)
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
pd.DataFrame(x_scaled,columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
x_train.size
```

```python
y_train.size
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=40,max_depth=1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
c=confusion_matrix(y_test,y_pred)
c
```

```python
plt.figure(plt.figure(figsize=(5,4)))
sns.heatmap(c,annot=True,fmt=".1f")
```
